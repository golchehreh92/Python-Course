# Distribution plot

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"], kde= False)    ##for removing the kde
```

```python
sns.distplot(tips["total_bill"], kde=False , bins=5)  ## we can increase or decrease the accuracy of the table by BINS
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=80)
```

# Joint plot

```python
sns.jointplot(x="total_bill" , y="tip" , data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")  #with kind we can change kind of the chart like hex or regression
```

# plot with limitation

```python
sns.jointplot(x="total_bill",y="tip",data=tips,xlim=(10,20),ylim=(0,4),kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex",color="green")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde",color="red")
```

```python
sns.jointplot(x="total_bill",y="tip",xlim=(0,18),ylim=(0,5),data=tips,kind="kde")
```

# Grids

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris["species"].nunique()
```

```python
iris["species"].unique()
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
p=sns.PairGrid(iris)
p.map_diag(sns.distplot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot,color="red")
```

```python
l=sns.FacetGrid(tips,row="time",col="smoker")
l.map(sns.distplot,"total_bill")
```
