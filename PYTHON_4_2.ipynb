```python
glare=df2.truncate(before="2021-03-23 08:00:00" , after="2021-03-23 17:00:00")

fig=plt.figure(figsize=(15,10))
axes=fig.add_axes([0.1,0.1,1,1])
axes.plot(glare["U1_Ev"],linestyle="--",label="U1",marker="o")
axes.plot(glare["U2_Ev"])
axes.plot(glare["U3_Ev"])
axes.plot(glare["U4_Ev"])
axes.plot(glare["U5_Ev"])

axes.set_xlabel("time step")
axes.set_ylabel("Lux")
axes.set_title("Ev")

plt.legend()
```
