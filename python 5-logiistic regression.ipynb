```python
X=pd.get_dummies(df)  ##این دستور هر ستون دارای استرینگ رو میکنه دو تا ستون و ارزش هاشو باینری میکنه 
```

```python
y=df["discomfort"]
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

# prediction

```python
prediction=logmodel.predict(X_test)
```

```python
prediction 
```

```python
logmodel.score(X,y)    ##bar asas train mast 
```

# Confusion matrix

```python
##confusion vase har model predisctioni k darim 
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,prediction))   ## about 53% we are accurate
```
