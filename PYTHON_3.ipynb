# Matplotib

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

# Functional method

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
```

# Object oriented method

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Lable")
axes.set_ylabel("Y Lable")
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.1,1,1])
axes2=fig.add_axes([.2,.5,.3,.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
op=plt.figure()
axes1=op.add_axes([.1,.1,2,2])
axes2=op.add_axes([.2,1.5,.4,.5])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("larger P")
axes2.set_title("smaller p")
```

```python
# object oriented is suitable for times we  want to have two fig together
```

```python
fig,axes=plt.subplots(1,3)
axes[0].plot(x,y)
axes[0].set_title("first")
axes[1].plot(y,x)
axes[1].set_title("second")
axes[2].plot(x,x)
axes[2].set_title("third")
```

```python
fig=plt.figure(figsize=(18,5),dpi=125)
axes=fig.add_axes([1,1,2,2])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(5,8),dpi=50)
axes[0].plot(x,y)
axes[1].plot(x,x)
axes[2].plot(y,y)
```

```python
fig,axes=plt.subplots(2,3)
axes[0][0].plot(x,y)
axes[0][1].plot(x,x)
axes[0][2].plot(y,x)
## inja hame tu ham tu ham an engar
```

```python
fig,axes=plt.subplots(2,3)
axes[0][0].plot(x,y)
axes[0][1].plot(x,x)
axes[0][2].plot(y,x)

plt.tight_layout()   ##farman 2rst shodan inha 
```

```python
fig=plt.figure(figsize=(3,4))
axes=fig.add_axes([1,1,2,2])
axes.plot(x,x**2,label="x square")
axes.plot(x,x**3,label="x cubic")
axes.legend()
```

# changing the color

```python
fig=plt.figure()
axes=fig.add_axes([1,1,2,2])
axes.plot(x,x)
axes.plot(x,y)
```

# changing the size and style of the pen

```python
fig=plt.figure(figsize=(3,4))
axes=fig.add_axes([1,1,2,2])
axes.plot(x,x**3, label="x", color="green", lw=4, linestyle="--")
axes.plot(y,y,label="y", color="pink", lw=3, linestyle="--")
```

# identifying the points exactly on the line

```python
fig=plt.figure(figsize=(3,3))
axes=fig.add_axes([1,1,2,2])
axes.plot(x,x**3, color="yellow", lw=2, linestyle="--", marker="o", markersize="7", markerfacecolor="green", markeredgecolor="black"
         ,markeredgewidth="2")
```
